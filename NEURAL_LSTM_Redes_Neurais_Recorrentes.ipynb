```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

ages = np.random.randint(low=15, high=70, size=40)

ages
```

```python
labels = []

for age in ages:
  if age < 30:
    labels.append(0)
  else:
    labels.append(1)

# RANDOM SWAP
for i in range(0, 3): # Sorteia 4 pessoas do conjunto
  r = np.random.randint(0, len(labels) - 1)
  if labels[r] == 0:
    labels[r] = 1
  else:
    labels[r] = 0
```

```python
plt.scatter(ages, labels, color="red")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(ages.reshape(-1,1), labels)
```

```python
# Equacao da RETA
# y = m.x + b

m = model.coef_[0]
b = model.intercept_

m,b
```

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

# marking the x-axis and y-axis
axis = plt.axes(xlim = (0,2), ylim=(-0.1,2))

# initializing a line variable
line, = axis.plot([], [], lw = 3)

# data which the line will contain (x,y)
def init():
  line.set_data([], [])
  return line,

def animate(i):
  m_copy = 1 * 0.01
  plt.title('n = ' + str(m_copy))
  x = np.arange(0.0, 10.0, 0.1)
  y = m_copy * x + b
  line.set_data(x, y)

  return line,

ani = FuncAnimation(fig, animate, init_func = init, frames = 200, interval = 20, blit = True)

ani.save('m.mp4', writer = 'ffmpeg', fps = 30)
```

```python
import math

def sigmoid(x):
  a = []
  for item in x:
    a.append(1/(1+math.exp(-item)))
  return a

x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)

plt.plot(x, sig)
plt.show()
```
